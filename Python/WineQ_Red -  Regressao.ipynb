----------------------------- Arvore de decisão para tarefa de regressão -------------------------------------------

```python
# escolha das features para o X foi feita olhando quais variaveis tem maior correlação com o alvo
# e evitei utilizar variaveis que possuem correlação forte com outras variaveis, como por exemplo evitei utilizar a "citric acid"
# por ter corelação forte com a "volatile acidity", que utilizarei pois possui correlação maior

x = DataWine[['alcohol','volatile acidity','sulphates','total sulfur dioxide']]
y = DataWine['quality']
```

```python
from sklearn.model_selection import train_test_split

x_treino,x_teste,y_treino,y_teste = train_test_split(x,y, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeRegressor

modelo_tree = DecisionTreeRegressor()

modelo_tree.fit(x_treino,y_treino)
```

```python
y_pred_teste  = modelo_tree.predict(x_teste)
y_pred_treino = modelo_tree.predict(x_treino)
```

```python
import sklearn.metrics as metrics
from math import sqrt
```

```python
MAE_treino = metrics.mean_absolute_error(y_treino,y_pred_treino)
R2_treino = metrics.r2_score(y_treino,y_pred_treino)
RMSE_treino = sqrt(metrics.mean_squared_error(y_treino,y_pred_treino))

print("---- Resultado Treino ---")
print(" ")
print("Mae : ", round(MAE_treino,2))
print("R2 : ", round(R2_treino,2))
print("RMSE : ", round(RMSE_treino,2))
```

```python
MAE_teste = metrics.mean_absolute_error(y_teste,y_pred_teste)
R2_teste = metrics.r2_score(y_teste,y_pred_teste)
RMSE_teste = sqrt(metrics.mean_squared_error(y_teste,y_pred_teste))

print("---- Resultado Teste ---")
print(" ")
print("Mae : ", round(MAE_teste,2))
print("R2 : ", round(R2_teste,2))
print("RMSE : ", round(RMSE_teste,2))
```

```python
print("qualidade média do vinho :" , round(y_teste.mean(),2))
print("Porcentagem do MAE em relação a media da base :", round(metrics.mean_absolute_error(y_teste, y_pred_teste)/y_teste.mean()*100,2))
```

----------------------------------------- Regressão Linear -------------------------------------------

```python
import statsmodels.api as sm
```

```python
x_treino_ = sm.add_constant(x_treino)
x_teste_  = sm.add_constant(x_teste)

Modelo_lr = sm.OLS(y_treino,x_treino_).fit()
```

```python
y_pred_treino_sm = Modelo_lr.predict(x_treino_)
y_pred_teste_sm  = Modelo_lr.predict(x_teste_)

Modelo_lr.summary()
```

```python
MAE_treino_sm = sm.tools.eval_measures.meanabs(y_treino,y_pred_treino_sm)
R2_treino_sm = Modelo_lr.rsquared
RMSE_treino_sm = sm.tools.eval_measures.rmse(y_treino,y_pred_treino_sm)

print("---- Resultado Treino SM ---")
print(" ")
print("Mae : ", round(MAE_treino_sm,2))
print("R2 : ", round(R2_treino_sm,2))
print("RMSE : ", round(RMSE_treino_sm,2))
```

```python
MAE_teste_sm = sm.tools.eval_measures.meanabs(y_teste,y_pred_teste_sm)
R2_teste_sm = Modelo_lr.rsquared
RMSE_teste_sm = sm.tools.eval_measures.rmse(y_teste,y_pred_teste_sm)

print("---- Resultado Teste SM ---")
print(" ")
print("Mae : ", round(MAE_teste_sm,2))
print("R2 : ", round(R2_teste_sm,2))
print("RMSE : ", round(RMSE_teste_sm,2))
```
