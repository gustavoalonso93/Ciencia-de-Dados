```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

```python
IrisData = pd.read_csv('iris.data')
IrisData.columns = ['Sepal Length','Sepal Width','Petal Length','Petal Width','Type']
```

```python
IrisData.hist(figsize=(20,15),
              bins=20)
```

```python
colores = {'Iris-virginica':"Blue",'Iris-setosa':"Orange",'Iris-versicolor':"Red" }
Array_Colors =[colores[x] for x in IrisData['Type']]
```

```python
plt.figure(figsize=(15,10))
plt.scatter(IrisData['Petal Length'],IrisData['Petal Width'], c=Array_Colors)
plt.xlabel("Petal Length")
plt.ylabel("Petal Width")
plt.title("Petal" , fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.scatter(IrisData['Sepal Length'],IrisData['Sepal Width'], c=Array_Colors)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Sepal" , fontsize=20)
plt.show()
```

```python
x = IrisData[['Petal Length','Petal Width']]
y = IrisData['Type']
```

```python
from sklearn.model_selection import train_test_split

x_treino,x_teste,y_treino, y_teste = train_test_split(x,y, test_size=0.3)
```

```python
from sklearn.tree import DecisionTreeClassifier

Decisiontree = DecisionTreeClassifier(criterion='gini')
```

```python
Decisiontree.fit(x_treino,y_treino)
```

```python
pred_teste = Decisiontree.predict(x_teste)
pred_treino = Decisiontree.predict(x_treino)
```

```python
from sklearn.metrics import confusion_matrix, classification_report

print(confusion_matrix(y_teste,pred_teste))
print("--------------------------")
print("Base de Teste")
print(classification_report(y_teste,pred_teste))
```

```python
print(confusion_matrix(y_treino,pred_treino))
print("--------------------------")
print("Base de Treino")
print(classification_report(y_treino,pred_treino))
```
